## Soal 2 (5 poin)
1. Lakukan proses data imputation pada fitur yang memiliki data yang hilang
2. Cek kembali apakah masih terdapat data yang hilang

Proses data imputation pada fitur yang memiliki data yang hilang

```python
# modus
df['occupation'].fillna(value=df['occupation'].mean, inplace=True)


df['workclass'].fillna(value="workclass", inplace=True)

# modus
df['native-country'].fillna(value=df['native-country'].mode, inplace=True)
```

```python
# Jawab 1.2 - Barchart
```

```python
num_cols = ['education']
for col in num_cols:
  plt.figure()
  plt.bar(df[col].value_counts().index, df[col].value_counts().values)
  plt.title(f'Distrubusi education')
  plt.xlabel('education'); plt.ylabel('Frekuensi')
  plt.xticks(rotation=60)
  plt.show()
```

```python
# Jawab 1.3 - Boxplot
```

```python
num_cols = ['hours-per-week']
for col in num_cols:
    plt.figure()
    df.boxplot(column='hours-per-week', by='income')
    plt.title(f'Boxplot income')
    plt.xlabel('income')
    plt.ylabel('hours-per-week')
    plt.show()
```

```python
# Jawab 1.4 - Lineplot
```

```python
mean_age = df.groupby("age")[["capital-gain", "capital-loss"]].mean()
plt.figure(figsize=(12, 6))
sns.lineplot(data=mean_age, x="age", y="capital-gain", label="Capital Gain")
sns.lineplot(data=mean_age, x="age", y="capital-loss", label="Capital Loss")

plt.title("Hubungan Age dengan Capital-Gain dan Capital-Loss")
plt.xlabel("Age")
plt.ylabel("Jumlah")
plt.legend()
plt.show()
```

# Bagian 3 - Encoding Variabel Kategorical

## Soal 1 (5 poin)
Lakukan encoding pada 'Sex' dan 'Income'. 'Income' merupakan variabel target

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['sex'] = le.fit_transform(df['sex'])
df['income'] = le.fit_transform(df['income'])
```

# Bagian 4 - Analisis Korelasi

## Soal 1 (10 poin)
1. Lakukan analisis korelasi pada variabel 'age', 'education-num', 'hours-per-week', 'capital-gain', 'capital-loss', dan 'income' (yang sudah di-encoding)
2. Berdasarkan hasil korelasi, informasi apa yang dapat Anda interpretasikan?

```python
import seaborn as sns #

cols = ['age', 'education-num', 'hours-per-week', 'capital-gain', 'capital-loss', 'income']
corr = df[cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, cmap='coolwarm', vmin=-1, vmax=1, annot=True, fmt=".2f")
plt.title('Analisis Korelasi')
plt.tight_layout()
plt.show()
```

```python
'''
Hasil analisisnya:
age dan income menunjukkan korelasi positif dimana semakin bertambah usia maka pengalaman lebih tinggi juga
'''
```
